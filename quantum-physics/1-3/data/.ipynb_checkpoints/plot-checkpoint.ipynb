```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
I, V = [], []
with open("./2.53.csv", 'r') as file:
  csvreader = csv.DictReader(file, delimiter=',')
  for row in csvreader:
    print(row)
    I.append(float(row['I']))
    V.append(float(row['V']))
```

```python
XErr = [0.01 for _ in V]
YErr = [0.001 for _ in I]
plt.figure(figsize=[16,9])
plt.scatter(V, I, color="red")
plt.errorbar(V, I, YErr, XErr)
plt.grid()
plt.title("ВАХ тиратрона ($U_h = 2.53\;V$)")
plt.xlabel("U, V")
plt.ylabel("I, mA")
plt.savefig("images/2.53.png")
```

```python
c = 1
w = -1/c * np.log(np.asarray(I))
WErr = 1/np.asarray(I) * YErr
plt.figure(figsize=[16,9])
plt.scatter(V, w, color="red")
plt.errorbar(V, w, WErr, XErr, ecolor="orange")
plt.grid()
plt.title("Качественный вид $\omega(V)$ при $U_h = 2.53\;V$")
plt.xlabel("V, V")
plt.ylabel("$\omega$")
plt.savefig("images/omega2.53.png")
```
