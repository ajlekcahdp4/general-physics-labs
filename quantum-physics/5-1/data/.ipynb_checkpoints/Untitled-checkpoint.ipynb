```python
import csv
import numpy
import matplotlib.pyplot as plt
```

```python
def readCSV(Filename):
    with open (Filename) as f:
        reader = csv.DictReader(f)
        Al, Fe, Pb =[], [], []
        for row in reader:
            Al.append(float (row['Al']))
            Fe.append(float(row['Fe']))
            Pb.append(float(row['Pb']))
        N = list(range(1, len(Al)+1))
        return N, Al, Fe, Pb
```

```python
N, Al, Fe, Pb = readCSV('plot.csv')
plt.figure(figsize=[16,9])
plt.title("Plot $\mu$ of different materials")
plt.xlabel("N")
plt.ylabel("$\mu, mm^{-1}$")
plt.scatter(N, Al)
plt.plot(N, Al)
plt.scatter(N, Fe)
plt.plot(N, Fe)
plt.scatter(N, Pb)
plt.plot(N, Pb)
```
